```python
# metoda Monte Carlo 
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-x**2)

a, b = -3, 3
os_x = np.linspace(start=a, stop=b, num=300)
os_y = f(os_x)
ymin, ymax = os_y.min(), os_y.max()
N = 1000
x = rnd.uniform(a, b, size=(1,N))
fx = f(x)
y = rnd.uniform(ymin, ymax, size=(1,N))

p_wew = y < fx 
x_wew = x[p_wew]
y_wew = y[p_wew]

fig, ax = plt.subplots()
ax.plot(os_x, f(os_x))
ax.plot(x, y, '.', color='gray')
ax.plot(x_wew, y_wew, '.', color='tab:orange')
plt.show()

licznik = x_wew.size / x.size
print(licznik*(b-a)*(ymax-ymin))
```

```python
import matplotlib.pyplot as plt

bramki = set(['A', 'B', 'C'])
wygrane = {'uparty':0, 'elastyczny':0}

for i in range(5000):
    bramka_wyg = set(rnd.choice(list(bramki)))
    bramki_przeg = bramki - bramka_wyg

    bramka_gracza = set(rnd.choice(list(bramki)))
    otwarta_bramka = set(rnd.choice(list(bramki_przeg - bramka_gracza)))
    decyzja_upartego = bramka_gracza
    decyzja_elastycznego = bramki - otwarta_bramka - bramka_gracza
    if decyzja_upartego == bramka_wyg:
        wygrane['uparty'] += 1
    if decyzja_elastycznego == bramka_wyg:
        wygrane['elastyczny'] += 1

print(wygrane)
print(wygrane['uparty']/(wygrane['uparty']+wygrane['elastyczny']))
plt.bar(list(wygrane.keys()), wygrane.values())
plt.show()
```
